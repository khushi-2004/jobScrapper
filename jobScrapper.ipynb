```python
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as bs
import logging
```

```python
intern_url='https://internshala.com/internships/'+'cyber-security'+'-internship/'
print(intern_url)
```

```python
urlClient=uReq(intern_url)
```

```python
# read the data

intern_page=urlClient.read()
```

```python
# this page is not human readable form so we buetyfy this
intern_html=bs(intern_page,'html.parser')
```

```python
bigbox=intern_html.find_all("div",{"class":"container-fluid individual_internship visibilityTrackerItem"})
```

```python
len(bigbox)
```

```python
l=[]
for i in range(len(bigbox)):
#     print("https://internshala.com/"+bigbox[i].h3.a['href'])
    l.append("https://internshala.com/"+bigbox[i].h3.a['href'])
    requests.get(l[i])
print(l)
```

```python
for i in range(len(l)):
    job_req=requests.get(l[i])
    job_html=bs(job_req.text,'html.parser')
    info_company=job_html.find_all("div",{"class":"company"})
    hiring_post=info_company[0].span.text
    print(hiring_post)
    
    company_name=info_company[0].a.text.strip()
    print(company_name)
    
    job_detail=job_html.find_all("div",{"id":"location_names"})
    job_location=job_detail[0].a.text
    print(job_location)
    print('\n')
    
    internship_detail=job_html.find_all("div",{"class":"internship_details"})
    detail=internship_detail[0].div.text
    print("About the internship: ",detail)
    print('\n')
    
    skill_required=job_html.find_all("div",{"class":"round_tabs_container"})
    skill_requires=skill_required[0].text.strip()
    print("required skills is: ",skill_requires)
    print("earn certificate in these skills :")
    print('\n')
    
    Earn_certifications=job_html.find_all("div",{"class":"training_link"})
    for i in range(len(Earn_certifications)):
        topic=Earn_certifications[i].a.text
        link=Earn_certifications[i].a['href']
        print(topic)
        print(link)
        print('\n')
    detail=job_html.find_all("div",{"class":"text-container"})
    l1=[]
    for i in range(len(detail)-1):
        l1.append(detail[i+1].text)
        print("who can apply : \n",l1[i])
        print("other requirment: \n",l1[i])
        print("aditional information: \n",l1[i])
        print("number of opening: \n",l1[i])
        print("about the company: \n",l1[i])
        
#         print(" \n",l1[i])
    print('\n\n')
```
